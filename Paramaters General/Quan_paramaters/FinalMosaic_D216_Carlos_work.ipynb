```python
import numpy as np,os,sys,glob
import matplotlib.pylab as plt
from ioMicro import *
```

```python
data_flds = glob.glob(r'P:\Adam\E216_WTC11_day14_WTnew\H4*_DMER_4*')
for data_fld in data_flds:
    resc = 4
    icol = -1
    if os.path.isdir(data_fld):
        fl_save = data_fld+'_col'+str(icol)+'.png'
        print(fl_save)
        if not os.path.exists(fl_save) or True:
            fls_ = np.sort(glob.glob(data_fld+r'\*.zarr'))
            ims,xs_um,ys_um=[],[],[]
            from tqdm.notebook import tqdm

            for fl in tqdm(fls_[:]):
                im,x,y = read_im(fl,return_pos=True)
                ims.append(np.array(im[icol][15,::resc,::resc],dtype=np.float32))
                xs_um.append(x)
                ys_um.append(y)

            im_big,xs,ys = compose_mosaic([im[::-1,::-1] for im in ims],xs_um,ys_um,ims_c=None,um_per_pix=0.108333*resc,
                                          rot = 0,return_coords= True)

            resc_ = 3
            data_fld = os.path.dirname(fls_[0])
            fig = plt.figure(figsize=(30,30))
            plt.imshow(im_big.T[::resc_,::resc_],vmin=0,vmax=np.percentile(im_big[::resc_,::resc_],99.),cmap='gray')

            
            fig.savefig(fl_save)
            for x_,y_,fl_ in zip(xs,ys,fls_):
                ifov = fl_.split('_')[-1].split('.')[0]
                plt.text(x_/resc_,y_/resc_,ifov,color='r')
            plt.xticks([])
            plt.yticks([])
            fl_save = fl_save.replace('.png','_annot.png')
            print(fl_save)
            fig.savefig(fl_save)
            plt.close('all')
```
